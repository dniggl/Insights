```python
# Importing the libraries.
import os
import tweepy as tw
import pandas as pd
import numpy as np

# Connecting to your Twitter Developer API's.
api_key= 'API Key Here' 
api_secret= 'API Secret Key Here'
access_token= 'Access Token Here'
access_token_secret= 'Access Token Secret Here'

# Authenticate the APIs
auth = tw.OAuthHandler(api_key, api_secret)
auth.set_access_token(access_token, access_token_secret)
api = tw.API(auth, wait_on_rate_limit=True)

# Define the search words and the date_since date as variables.
search_words = "LA rams"
date_since = "2022-01-01"

# Collect the tweets.
tweets = tw.Cursor(api.search,
              q = search_words,
              lang = "en",
              since = date_since).items(5)

# Define a function to get the related tweets.
def get_related_tweets(key_word):
    twitter_users = []
    tweet_time = []
    tweet_string = [] 
    for tweet in tw.Cursor(api.search,q=key_word, count=1000).items(1000):
            if (not tweet.retweeted) and ('RT @' not in tweet.text):
                if tweet.lang == "en":
                    twitter_users.append(tweet.user.name)
                    tweet_time.append(tweet.created_at)
                    tweet_string.append(tweet.text)
                    
    df = pd.DataFrame({'name':twitter_users, 'time': tweet_time, 'tweet': tweet_string})
    df.to_csv(f"{key_word}.csv")
    return df

# Get tweets related to LA Rams and create a new dataframe.
df = get_related_tweets("LA Rams")
df.head(7)
```
