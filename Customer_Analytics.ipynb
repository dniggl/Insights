```python
# Import software libraries.
import pandas as pd
import matplotlib.pyplot as plt
import matplotlib.gridspec as gridspec
import seaborn as sns
import numpy as np
```

```python
# Import the dataset from a website.
data_source_url = "https://raw.githubusercontent.com/Jotaherrer/DataAnalysis/master/e_commerce/e_commerce.csv"
ecommerce = pd.read_csv(data_source_url)
```

```python
# Print the first 5 rows
ecommerce.head(50)  
```

The e_commerce data set was obtained from GitHub.

Feature variables include:
1.   User ID - Unique user identifier.
2.   Product_ID - Unique product identifier.
3.   Gender - Gender of customer (male or female).
4.   Age - Age category (0-17, 18-25, 26-35, 36-45, 46-50, 51-55, 55+).  
5.   Occupation - Occupation of customer. 
6.   City_Category - Area classification of city.  
7.   Stay_In_Current_City_Years - Number of years in current city.
8.   Marital_Status - Single or married. 
9.   Product_Category_1 - Purchased 1 product.
10.  Product_Category_2 - Purchased 2 products.
11.  Product_Category_3 - Purchased 3 or more products.
12.  Purchase - Dollar amount of purchase.


```python
# Show information about the data columns.
ecommerce.info()
```

There are 550,068 customer transactions originating from ecommerce platforms and 12 columns. The data set contains integer, object and float values. 

```python
# Check for NA or missing values in features.
ecommerce.isna().sum()
```

Product_Category_2 variable has 173,638 null values and Product_Category_3 variable has 383,247 null values.     

In order to clean the data, we will replace the null values with zeros. 

```python
# Replace the null values with zero (0).
ecommerce.fillna(0, inplace=True)
```

The dataframe includes every transacion made by each customer. Let's identify the users that spend the most money. 

```python
# Sum the purchase amounts by user id and show the users that have spent the most money on products.  
purchases = ecommerce.groupby(['User_ID']).sum().reset_index()
purchases = purchases.sort_values('Purchase',ascending=False)
purchases.head()
```

```python
# Visualize the distribution of sales by gender.
plt.figure(figsize=(8,5))
ecommerce.Gender.value_counts().plot(kind='pie', autopct='%1.0f%%')
plt.axis('equal')
plt.title("Gender Distribution", fontsize='18')
plt.show()
```

```python
# Visualize the distribution of sales by age groups.
plt.figure(figsize=(8,5))
ecommerce.Age.value_counts().plot(kind='pie', autopct='%1.0f%%')
plt.axis('equal')
plt.title("Age Distribution", fontsize='18')
plt.show()
```

Find the average sale amount for each age group.

```python
purchase_by_age = ecommerce.groupby('Age')['Purchase'].mean().reset_index()
purchase_by_age
```

The age group 5155 spends the most amount on products. 

```python
# Visualize the distribution of average purchase amounts for each age group.
plt.figure(figsize=(10,4))
plt.plot(purchase_by_age.Age, purchase_by_age.Purchase, color='darkblue', marker='*')
plt.grid()
plt.xlabel('Age Group', fontsize=10)
plt.ylabel('Total Purchases in Dollars', fontsize=10)
plt.title('Average Sales by Age Group', fontsize=13)
plt.show()
```

```python
# Visualize purchases by occupation.
# Group by occupation.
occupation = ecommerce.groupby('Occupation')['Purchase'].mean().reset_index()

# Plot bar chart for average purchases by occcupation.
sns.set(style="white", rc={"lines.linewidth": 3})
fig, ax1 = plt.subplots(figsize=(10,6))
sns.barplot(x=occupation['Occupation'],y=occupation['Purchase'],color='#004488',ax=ax1)
plt.axis([-1,21,8000,10000])
plt.title('Average Purchases by Occupation', fontsize='14')
plt.show()
```

```python
# Visualize the best selling products. 
# Group by product ID
product = ecommerce.groupby('Product_ID')['Purchase'].count().reset_index()
product.rename(columns={'Purchase':'Count'},inplace=True)
product_sorted = product.sort_values('Count',ascending=False)

# Plot line plot
plt.figure(figsize=(12,7))
plt.plot(product_sorted['Product_ID'][:10], product_sorted['Count'][:10], linestyle='-', color='darkblue', marker='o')
plt.title("Best-selling Products", fontsize='15')
plt.xlabel('Product ID', fontsize='15')
plt.ylabel('Products Sold', fontsize='15')
plt.show()
```

Summary of findings include the following:

* The largest amount spent on products was $10,536,909.
* 75% of purchases were made by males and 25% of purchases were made by females.
* Age group 26-35 made 40% of the total purchases followed by age group 36-45 who made 20% of the total purchases. 
* On average, age group 5155 spends the most money on products followed by age group 55+.  
* On average, the occupations of the customers that spend the most money include numbers 12, 15 and 17. Unfortunately there were no translations  for occupation numbers to specify the type of occupation.        
* The best selling products were P00265242, POOO25442 AND P00110742. Unfortunately there were no translations for product numbers to specify the type of product.
